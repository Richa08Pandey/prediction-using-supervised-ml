# import necessary libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python

data = pd.read_csv('student_scores.csv') # reading DataSet
data.head()  #check top 5 tuples
```

# Read the csv dataset as a pandas dataframe

```python
data.info() #get information about DataSet like memory usage,datatype...
```

```python
data.describe() #description of data in dataframe
```

# Visualize the data

```python
sns.scatterplot(x=data['Hours'], y=data['Scores']);  #here we are ploting data
```

```python
sns.regplot(x=data['Hours'], y=data['Scores']);  #regression plot gives better visualization 
```

# Separate feature(s) and target

```python
X = data[['Hours']]
Y = data['Scores']
```

```python
# train-test split
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_Y,val_Y = train_test_split(X,Y,random_state = 0)
```

# Model building

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(train_X,train_Y) #train the model
```

```python
pred_Y = regressor.predict(val_X) #making prediction
```

```python
pd.DataFrame({'Actual': val_Y, 'Predicted': pred_Y})  
```

```python
## Actual Vs Predicted distribution plot

sns.kdeplot(pred_Y,label="Predicted", shade=True);

sns.kdeplot(data=val_Y, label="Actual", shade=True);
```

```python
print('Train accuracy: ', regressor.score(train_X,train_Y),'\nTest accuracy : ',regressor.score(val_X,val_Y))
```

```python
h = [[9.25]]
s = regressor.predict(h)
print('A student who studies ',h[0][0],'hours is estimsted to score ', s[0])
```
